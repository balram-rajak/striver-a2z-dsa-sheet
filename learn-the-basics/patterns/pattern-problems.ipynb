````python
def printSquare(n):
    for i in range(n):
        for j in range(n):
            print("*", end=" ")
        print()

printSquare(5)
````

````python
def printStarTriangle(n):
    for i in range(n):
        for _ in range(i+1):
            print("*",end=" ")
        print()
````

````python
def printStarTriangleShort(n):
    for i in range(n):
        print("* " * (i+1))   
````

````python
def reverseTriangle(n):
    for i in range(n,0,-1):
        print("* " * (i))

reverseTriangle(5)
````

````python
def reverseTriangleNumbers(n):
    for i in range(n,0,-1):
        for j in range(i):
            print(j+1,end=" ")
        print()

reverseTriangleNumbers(5)
````

````python
def equilateralTriangle(n):
    for i in range(n+1):
        for j in range(1,2*n):
            if(j < n-i+1 or j > n+i-1):
                print(end=" ")
            else:
                print("*",end="")
        print()
print(equilateralTriangle(5))
````

````python
def equilateralTriangleReverse(n):
    for i in range(n,0,-1):
        for j in range(2*(n)-1,0,-1):
            if(j < n-i+1 or j > n+i-1):
                print(end=" ")
            else:
                print("*",end="")
        print()
print(equilateralTriangleReverse(5))
````

````python
def parallelogram(n):
    for i in range(n+1):
        for j in range(1,2*n):
            if(j < n-i+1 or j > n+i-1):
                print(end=" ")
            else:
                print("*",end="")
        print()
    for i in range(n,0,-1):
        for j in range(2*(n)-1,0,-1):
            if(j < n-i+1 or j > n+i-1):
                print(end=" ")
            else:
                print("*",end="")
        print()
    
print(parallelogram(5))
````

````python
def halfBigDiamond(n):
    for i in range(n-1):
        print("* " * (i+1)) 
    print("* " * n)
    for i in range(n-1,0,-1):
        print("* " * (i))

print(halfBigDiamond(5))
````

````python
def p_19(x):
    for i in range(x):
        chars = "* "*(x-i);
        print(chars+" "*2*(i)+chars)
    for i in range(x):
        chars = "*"*(i+1);
        print(chars+" "*2*(x-i-1)+chars)
print(p_19(5))
````

````python
def p_20(x: int):
    
    for i in range(2*x-1):
        
        if(i<=x-1):
            space = (x-i-1);
        elif(i>x-1):
            space = i%x+1
            
        stars = "* "*(x-space);
        print(stars,end="");
        print("    "*space,end="");
        print(stars)

p_20(3)
````
